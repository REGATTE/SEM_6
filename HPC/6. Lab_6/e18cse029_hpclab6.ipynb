```
import random
import numpy as np
samay = []

n = int(input())
mat = np.random.randint(low = 0, high = 100, size = (n,n))
new_mat = np.random.randint(low = 0, high = 1, size = (n,n))
import time
start = time.time()
for i in range(n):
  for j in range(n):
    for k in range(n):
      new_mat[i][j]+= mat[i][k]*mat[k][j]
end = time.time()
samay.append(end-start)
print("matrix")
print(mat)
print("naya Matrix")
print(new_mat)
```

```
import random
import numpy as np
n = int(input())
mat = np.random.randint(low = 0, high = 100, size = (n,n))
start = time.time()
new_mat = np.matmul(np.matrix(mat),np.matrix(mat))
end = time.time()
print(new_mat)
samay.append(end-start)
print(end-start)
```

```
from threading import Thread
def multiply(a,b, mat, new_mat):
  for i in range(a,b):
    for j in range(n):
      for k in range(n):
        new_mat[i][j]+= mat[i][k]*mat[k][j]
threads = [None] * (n)
start = time.time()
for i in range(len(threads)):
    threads[i] = Thread(target=multiply, args=(i,i+1,mat,new_mat))
    threads[i].start()

for i in range(len(threads)):
    threads[i].join()

# print
end = time.time()
samay.append(end-start)
print(new_mat)
```
