using multithreading

```
%%time
import threading 
import os
import numpy as np 
n=int(input())
A = [[0 for i in range()]  
            for j in range(n)] 
B = [[0 for i in range(n)]  
            for j in range(n)]
print("Enter the elements of First Matrix: ")  
for i in range(n) :  
    for j in range(n) :  
            A[i][j] = int(input("A[" + str(i) +
                                "][" + str(j) + "]: "))
print("Enter the elements of First Matrix: ")  
for i in range(n) :  
      for j in range(n) :  
            B[i][j] = int(input("B[" + str(i) +
                                "][" + str(j) + "]: "))           
def fun(mat1, mat2, fin, i):
  for j in range(n):
    for k in range(n):
      fin[i][j] += mat1[i][k] * mat2[k][j]


threads = [None]*n
fin = np.zeros((n,n))

for i in range(n):
  threads[i] = threading.Thread(target = fun, args = (A, B, fin, i))
  threads[i].start()

for i in range(n):
  threads[i].join()

print(fin)
```

```
%%time
a = np.random.randint(100,size = 1000000)
def  sum(a):
    sum = 0
    for i in range(len(a)):
        sum += a[i]
    return sum

qu = queue.Queue()
thread_list = list()

l=a.shape[0]
final_sum = 0

t0 = threading.Thread(target=lambda b, arg1: b.put(sum(arg1)), args=(qu, a[:(l//2)]))
t0.start()
threads_list.append(t0)

t1 = threading.Thread(target=lambda b, arg1: b.put(sum(arg1)), args=(qu, a[(l//2):]))
t1.start()
thread_list.append(t1)

for t in threads_list:
    t.join()

while not qu.empty():
    final_sum += qu.get()
print(final_sum)
```

using threading


```
%%time

def prime(l0,hi,res):
    result=[]
    for num in range(l0, hi + 1):
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           result.append(num)
    res.put(result)

q = queue.Queue()
thread_list = list()

final_res = []

t = threading.Thread(prime(lo, hi//2, q))
t.start()
thread_list.append(t)

t0 = threading.Thread(prime(hi//2, hi, q))
t0.start()
thread_list.append(t0)

for x in threads_list:
    x.join()

for i in q.queue:
    final_res +=i
print(final_res)
```
